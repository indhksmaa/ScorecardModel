```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import scipy
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
train = pd.read_csv('train_cleaned.csv')
```

```python

sns.histplot(data=train, x='DAYS_EMPLOYED', hue='TARGET', multiple='stack',kde=True)
plt.xlabel('Days Employed')
plt.ylabel('Frequency')
plt.title('Effect of Days Employed on Payment Difficulties')
```

```python

sns.histplot(data=train, x='DAYS_LAST_PHONE_CHANGE', hue='TARGET', multiple='stack',kde=True)
plt.xlabel('DAYS LAST PHONE CHANGE')
plt.ylabel('Frequency')
plt.title('Effect of Days Last Phone Change on Payment Difficulties')
```
